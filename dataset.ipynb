```python
import cv2
import os

# Input and output directories
input_dir = "/media/user/9c7eaef1-35fa-4210-889c-9e2b99342586/user/abul/dataset sdo patches low res"  # Change this to your actual folder
output_dir = "/media/user/9c7eaef1-35fa-4210-889c-9e2b99342586/user/abul/dataset sdo patches low res final"  # Change this to where you want to save

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Process all image files in the input directory
for filename in os.listdir(input_dir):
    if filename.lower().endswith((".jpg", ".png", ".jpeg", ".tif", ".bmp")):
        input_path = os.path.join(input_dir, filename)
        output_path = os.path.join(output_dir, filename)
        # Load the image as grayscale
        image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
        if image is not None:
            # Save the grayscale image
            cv2.imwrite(output_path, image)
            print(f"Converted and saved: {output_path}")
        else:
            print(f"Error loading: {input_path}")

print("Conversion complete!")
```
